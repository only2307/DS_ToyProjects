```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
epl_df = pd.read_csv("EPL_20_21.csv")
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
# count null values
epl_df.isna().sum()
```

```python
# create 2 new cols
epl_df["MinsPerMatch"] = (epl_df["Mins"]/epl_df["Matches"]).astype(int)
epl_df["GoalsPerMatch"] = (epl_df["Goals"]/epl_df["Matches"]).astype(float)
epl_df.head()
```

```python
# Total goals
total_goals = epl_df["Goals"].sum()
print(total_goals)
```

```python
# Penalty goals
total_penalty_goals = epl_df["Penalty_Goals"].sum()
print(total_penalty_goals)
```

```python
# Penalty attempts
total_penalty_attempts = epl_df["Penalty_Attempted"].sum()
print(total_penalty_attempts)
```

```python
# Pie chart for penalties missed vs scored
plt.figure(figsize=(13,6))
pl_missed = epl_df["Penalty_Attempted"].sum() - total_penalty_goals
data = [pl_missed, total_penalty_goals]
labels = ["Penalties missed", "Penalties scored"]
color = sns.color_palette("Set2")
plt.pie(data, labels=labels, colors=color, autopct="%.1f%%")
plt.show()
```

```python
# Unique position
epl_df["Position"].unique()
```

```python
# Number of FW players
epl_df[epl_df["Position"] == "FW"]
```

```python
# Players from different nations
np.size((epl_df["Nationality"].unique()))
```

```python
# Country with the most players
nationality = epl_df.groupby("Nationality").size().sort_values(ascending=False)
nationality.head(10).plot(kind="bar", figsize=(12,6), color=sns.color_palette("magma"))
```

```python
# Clubs with max players in their squads
epl_df["Club"].value_counts().nlargest(5).plot(kind="bar", color=sns.color_palette("viridis"))
```

```python
# Clubs with mininum players in their squads
epl_df["Club"].value_counts().nsmallest(5).plot(kind="bar", color=sns.color_palette("viridis"))
```

```python
# Players based on age group
under21 = epl_df[epl_df["Age"] <= 20]
age21_25 = epl_df[(epl_df["Age"] > 20) & (epl_df["Age"] <= 25)]
age26_30 = epl_df[(epl_df["Age"] > 25) & (epl_df["Age"] <= 30)]
above30 = epl_df[epl_df["Age"] > 30]
```

```python
x = np.array([under21["Name"].count(), age21_25["Name"].count(), age26_30["Name"].count(), above30["Name"].count()])
myLabels = ["<=20", "21 to 25", "26 to 30", ">30"]
plt.title("Total Players by Age", fontsize=28)
plt.pie(x, labels=myLabels, autopct="%.2f%%")
plt.show()
```

```python
# Total under 20 players in each club
players_under_20 = epl_df[epl_df["Age"] < 20]
players_under_20["Club"].value_counts().plot(kind="bar", color=sns.color_palette("cubehelix"))
```

```python
#Under 20 players in Manchester United
players_under_20[players_under_20["Club"] == "Manchester United"]
```

```python
#Under 20 players in Chelsea
players_under_20[players_under_20["Club"] == "Chelsea"]
```

```python
# Average age of players in each club
plt.figure(figsize=(12,6))
sns.boxplot(x="Club", y="Age", data=epl_df)
plt.xticks(rotation=90)
```

```python
num_players = epl_df.groupby("Club").size()
data = (epl_df.groupby("Club")["Age"].sum()/num_players)
data.sort_values(ascending=False)
```

```python
# Total assists from each club
assists_by_club = pd.DataFrame(epl_df.groupby("Club", as_index=False)["Assists"].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x="Club", y="Assists", data=assists_by_club.sort_values(by="Assists"), palette="Set2")
ax.set_xlabel("Club", fontsize=30)
ax.set_ylabel("Assists", fontsize=20)
plt.xticks(rotation=60)
plt.rcParams["figure.figsize"] = (20, 8)
plt.title("Club vs Total Assists", fontsize=30)
```

```python
# Top 10 assists
top_10_assists = epl_df[["Name", "Club", "Assists", "Matches"]].nlargest(n=10, columns="Assists")
top_10_assists
```

```python
# Total assists from each club
goals_by_club = pd.DataFrame(epl_df.groupby("Club", as_index=False)["Goals"].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x="Club", y="Goals", data=goals_by_club.sort_values(by="Goals"), palette="rocket")
ax.set_xlabel("Club", fontsize=30)
ax.set_ylabel("Goals", fontsize=20)
plt.xticks(rotation=60)
plt.rcParams["figure.figsize"] = (20, 8)
plt.title("Club vs Total Goals", fontsize=30)
```

```python
# Most goals by players
top_10_goals = epl_df[["Name", "Nationality", "Club", "Goals", "Matches"]].nlargest(n=10, columns="Goals")
top_10_goals
```

```python
# Goals per match
top_10_goals_per_match = epl_df[["Name", "Nationality", "Club", "Goals", "GoalsPerMatch"]].nlargest(n=10, columns="GoalsPerMatch")
top_10_goals_per_match
```

```python
# Pie chart - Goals vs Assists and Without assists
plt.figure(figsize=(14,7))
assists = epl_df["Assists"].sum()
data = [total_goals - assists, assists]
labels = ["Goals without Assists", "Goals with Assists"]
color = sns.color_palette("Set1")
plt.pie(data, labels=labels, colors=color, autopct="%.2f%%")
plt.show()
```

```python
# Top 10 players receiving most yellow cards
epl_yellow_cards = epl_df.sort_values(by="Yellow_Cards", ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Players with the most yellow cards")
c = sns.barplot(x=epl_yellow_cards["Name"], y=epl_yellow_cards["Yellow_Cards"], label="Players", color="yellow")
plt.ylabel("No. of yellow cards")
c.set_xticklabels(c.get_xticklabels(), rotation=45)
c
```
